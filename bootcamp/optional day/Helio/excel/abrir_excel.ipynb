```python
import pandas as pd

xlsx_df = pd.ExcelFile('resultado transporte 2017.1.xlsx')
me_interessam_df = []

for sheet in xlsx_df.sheet_names:
    planilha_df = pd.read_excel('resultado transporte 2017.1.xlsx', sheet_name=f'{sheet}')
    if len(planilha_df.columns) > 1:
        planilha_df = planilha_df.dropna()
        planilha_df.columns = planilha_df.iloc[0]
        planilha_df = planilha_df.iloc[1:]
        me_interessam_df.append(planilha_df)
grand_finalle_df = pd.concat(me_interessam_df)
grand_finalle_df.index = range(grand_finalle_df.shape[0])
grand_finalle_df
```
